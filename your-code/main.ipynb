#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.arange(0,100)
y = x**2
y= np.exp(x)
```

```python
f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize= (12,8))

ax1.plot(x,x**2, 'b--', x, np.exp(x), 'purple')
ax1.set_xscale('linear')
ax1.set_yscale('linear')
ax1.set_xlabel("x")
ax1.set_ylabel("x**2")

ax2.plot(x,x**2, 'b--', x, np.exp(x), 'purple')
ax2.set_yscale('log')
ax2.set_yscale('log')
ax2.set_xlabel("x")
ax2.set_ylabel("np.exp(x)")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
y_axis = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]
color_types = ["green", "purple", "red"]
for y,c in zip(y_axis,color_types):
    sns.relplot(data = fitbit, x = "Steps", y = y, color = c)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (12,4))
y_axis = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]
color_types = ["green", "purple", "red"]
axes = [ax1,ax2, ax3]

for y,c,a in zip(y_axis,color_types,axes):
    sns.scatterplot(data = fitbit, ax = a, x = "Steps", y = y, color = c)
```
